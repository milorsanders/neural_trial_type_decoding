### Overview
Parviz Ghaderi from Carl Petersen's lab at EPFL recorded the behavior and neural activity of mice in response to different stimuli. The experiments were divided into different sessions (with different mice on different days), with each session consisting of multiple trials. Each trial lasted 3 seconds and consisted of different combinations of stimuli (auditory tone or whisker deflection) provided to the mouse, and a potential lick response from the mouse. The mouse received a small reward whenever it licked in response to a certain combination of stimuli, and no reward otherwise. For this project, Tâm Nguyen prepared the raw experimental data so that we can try to read the minds of the mice.

*Goal: Given the average neural activity in 100ms time bins of different kinds of neurons in different brain areas, we want to predict the trial type, i.e., the perceived stimuli and the lick response. (auditory + whisker + lick response)*

### Background
Mouse neurons were recorded in **5 different brain areas**:

1. **A1** - Primary Auditory Cortex: Processes auditory information (tone stimuli)
2. **ALM** - Anterior Lateral Motor Cortex: Involved in motor planning and execution (licking behavior)
3. **wS1** - Primary Whisker Somatosensory Cortex: Processes whisker tactile information
4. **wS2** - Secondary Whisker Somatosensory Cortex: Higher-order whisker sensory processing
5. **wM2** - Secondary Whisker Motor Cortex: Motor control related to whisker movements

For each brain region, recordings were made from:
- **Multiple cortical layers**: L1, L2/3, L4, and L5 (varies by region)
- **Two neuron types**: 
  - **EXC** (Excitatory neurons): Increase activity in downstream neurons
  - **INH** (Inhibitory neurons): Decrease activity in downstream neurons

This results in **46 unique region-layer-celltype combinations**:
- A1: 10 combinations (L1, L4, L5, L6a, L6b, × 2 cell types)
- ALM: 6 combinations (L1, L2/3, L5 × 2 cell types)
- wM2: 8 combinations (L1, L2/3, L5, L6a × 2 cell types)
- wS1: 12 combinations (L1, L2/3, L4, L5, L6a, L6b × 2 cell types)
- wS2: 10 combinations (L1, L2/3, L4, L5, L6a × 2 cell types)

Neural activity for each combination was measured across **30 time bins**.

### Experiment
- Structure: **sessions → trials (3s) → stimulus/response**
- Stimuli:
  - **Auditory**: tones
  - **Whisker**: deflection
- Behavioral response: **lick**
- Reward delivered when **lick response** matches the required contingency.

### Training Data
The training set contains trials where the trial type is known. The trial type will be predicted for the test set.
The file `sample_submission.csv` is an example of how the submission will look like.

**Files**
- **train.csv**: the training set  
- **test.csv**: the test set  
- **sample_submission.csv**: a sample submission file in the correct format  

**Columns**
- **session_id**: mouse and day on which the recording was done  
- **trial_number**: trial number within the current session  
- **[Brain area]_[neuron type]_[time bin]**: average number of spikes of neurons in a given brain and given 100 ms time bin (e.g., `ALM_L1_EXC_time_13` contains the average number of spikes of excitatory neurons in layer 1 of the ALM area during 1200–1300 ms relative to recording onset).
- **TRIAL_TYPE**: response variable composed of trial type (what the mouse perceives) and response (what the mouse does), e.g. `NOGO W+ nolick` means that the no-Go cue was presented, the whisker was deflected, and the mouse did not lick.
    - **Auditory**: `GO`, `NOGO`, `no tone`
    - **Whisker**: `W+`, `W−`
    - **Lick response**: `lick`, `nolick`
